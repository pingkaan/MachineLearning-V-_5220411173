```python
import pandas as pd
dataku = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/Dataset/cuaca.xlsx')
```

```python
print("Ukuran:", dataku.shape)
```

```python
print("InFeatures:", dataku.columns)
```

```python
dataku
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in dataku.columns:
  dataku[col] = le.fit_transform(dataku[col])

dataku
```

```python
X = dataku[dataku.columns[:-1]]
y = dataku[dataku.columns[-1]]
```

```python
print("\nFeature matrix: \n", X)
print("\Response vector: \n", y)
```

```python
from sklearn.model_selection import train_test_split
X_latih, X_tes, y_latih, y_tes = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# pelatihan pada data latih menggunakan KNN (k=3)
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_latih, y_latih)
```

```python
# melakukan prediksi pada data uji
y_prediksi = knn.predict(X_tes)
```

```python
# perbandingan nilai aktual (y_tes) dengan nilai prediksi (y _prediksi)
from sklearn import metrics
print ("Akurasi model kNN:", metrics.accuracy_score(y_tes, y_prediksi))
```
